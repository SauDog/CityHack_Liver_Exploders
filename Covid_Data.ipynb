## Covid Data

```python
import pandas as pd
import numpy as np
pd.options.display.max_columns = 50
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
df = pd.read_csv('Covid Dataset.csv')
```

```python
df.head()
```

```python
df.replace('Yes', 1, inplace = True)
df.replace('No', 0, inplace = True)
```

```python
df.head()
```

```python
result = df['COVID-19']
symptoms = df.drop('COVID-19', axis = 1)
features = symptoms.columns.tolist()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(symptoms, result, test_size=0.2, shuffle = True, random_state = 69)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix

symptoms_prediction = model.predict(X_test)

#[Accurate    FalsePositve]
#[False Negative  Accurate]

print(confusion_matrix(y_test, symptoms_prediction))
```

```python
fig = plt.figure(figsize=(20,15))
_ = tree.plot_tree(model, feature_names = features, max_depth = 2)
```
